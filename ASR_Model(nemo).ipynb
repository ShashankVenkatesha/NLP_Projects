```python
import nemo.collections.asr as nemo_asr
List=[]
```

```python
asr_model_2 = nemo_asr.models.EncDecRNNTBPEModel.from_pretrained(model_name="stt_en_conformer_transducer_xlarge")
files = ['/content/No_4.mp3']
for fname, transcription in zip(files, asr_model_2.transcribe(paths2audio_files=files)):
  print(f"Audio in {fname} was recognized as: {transcription}")
List=transcription

```

```python
wc=1

print(List)
for i in List:
  for j in i:
    if j==" ":
      wc+=1
print(wc)  
```

```python
# METHOD 2
import audioread

# function to convert the information into
# some readable format
def duration_detector(length):
	
	mins = length /60 # calculate in minutes
	return mins


# alarm.wav is the name of the audio file
# f is the fileobject being created
with audioread.audio_open('No_4.mp3') as f:
	
	# totalsec contains the length in float
	totalsec = f.duration
	mins= duration_detector(int(totalsec))
	print(mins)

```

```python
rate=wc/mins
print("word count",wc)
print("the rate of speech for the given file is--->",rate)
```
